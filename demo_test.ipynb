```python
import pandas as pd
import re
```

```python
df_hours = pd.read_csv('different_cur_rate.csv')
df_currency_rate = pd.read_csv('currency_rates.csv')
```

```python
def clean_rate(rate:str)->float:

    regex_pattern = r"[^\d.]+"
    rate_str = re.sub(regex_pattern,"",str(rate))
    return float(rate_str)

def get_currency(rate:str)->str:

    regex_pattern = r"[^\d.]+"
    rate_str = re.findall(regex_pattern,rate)
    return rate_str[0].strip()
```

```python
test_cur = '$ 13.45'

get_currency(test_cur)
```
